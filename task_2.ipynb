```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
titanic_data = pd.read_csv("Titanic.csv")  # Replace with the actual file path

# Data Cleaning
titanic_data['Age'].fillna(titanic_data['Age'].median(), inplace=True)
titanic_data['Embarked'].fillna(titanic_data['Embarked'].mode()[0], inplace=True)

# Example Analysis: Survival rate by gender
gender_survival = titanic_data.groupby("Sex")["Survived"].mean()

# Visualization
sns.countplot(data=titanic_data, x='Pclass', hue='Survived')
plt.title("Survival Count by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.show()
```
